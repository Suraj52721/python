```python
import matplotlib.pyplot as plt
import numpy as np

def f(x):
  return np.cos(x)-x
def derivative(x):
  return -np.sin(x)-1

true_root = 0.73908

def newton(x0, tol, maxiter):
  x = x0
  for i in range(maxiter):
    x = x - f(x)/derivative(x)
    print("Iteration: ", i+1, "x = ", x, 'error', abs(f(x)-true_root))
    if abs(f(x)) < tol:
      break
  print("The root is: ", x)
  plt.plot(x, f(x), marker='o', color='r')
  
newton(1.5, 0.0001, 100)
print('\n')
newton(0.1, 0.0001, 100)
print('\n')
newton(-2, 0.0001, 100)

x = np.linspace(-2, 2, 20)
plt.plot(x, f(x))
plt.grid()
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.title('Graph of f(x) = cos(x)-x')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```
