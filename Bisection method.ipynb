```python
import matplotlib.pyplot as plt
import numpy as np
def f(x):
    return x**3-4*x-8

xn=[]
en=[]
true_val=2.649

def bisect(a,b,tol):
    m=np.linspace(a,b,100)
    plt.plot(m,f(m))
    if f(a)*f(b)<0:
        while (b-a)/2>tol:
            c=(b+a)/2
            if f(c)==0:
                xn.append(c)
                en.append(abs(c-true_val))
                print('the root is ',c)
            else:
                if f(a)*f(c)<0:
                    xn.append(c)
                    en.append(abs(c-true_val))
                    print('f(a)=',f(a),'f(c)=',f(c),'f(b)=',f(b))
                    b=c
                else:
                    xn.append(c)
                    en.append(abs(c-true_val))
                    print('f(a)=',f(a),'f(c)=',f(c),'f(b)=',f(b))
                    a=c
        print('The root of the equation is ',c)
        plt.plot(c,0,'*')
    else:
        print('Root doesn\'t exists')

bisect(1,5,0.0001)
plt.axhline(0,color='black')
plt.axvline(0,color='black')
plt.title('Graph of x³-4x-8')
plt.show()

print('xn=',xn)
print('en=',en)
em=[]
em.extend(en)
em.pop(0)
en.pop()
print('pop en=',en)
print('pop em=',em)
plt.plot(en,em)
plt.show
```
