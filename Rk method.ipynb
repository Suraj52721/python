```python
import math
import matplotlib.pyplot as plt
g=9.8
c=12.5
m=68.1

def f(t0,v0):
  return g-(c*v0/m)

def analytic(t):
  return (m*g/c)*(1-math.exp(-(c/m)*t))

error1=[]
def rk4(h,iteration):
  t0=0
  v0=0
  print(f'RK4 Method')
  for a in range(iteration):
    k1=h*f(t0,v0)
    k2=h*f(t0+h/2,v0+k1/2)
    k3=h*f(t0+h/2,v0+k2/2)
    k4=h*f(t0+h,v0+k3)
    vn=v0+(k1+2*k2+2*k3+k4)/6
    print(f't={a} v={v0} analytic={analytic(t0)}')
    t0+=h
    v0=vn
    error1.append(abs(analytic(t0)-v0))

rk4(2,25)

error2=[]
def rk2(h,iteration):
    t0=0
    v0=0
    print(f'RK2 Method')
    for a in range(iteration):
        k1=h*f(t0,v0)
        k2=h*f(t0+h,v0+k1)
        vn=v0+(k1+k2)/2
        print(f't={a} v={v0}  analytic={analytic(t0)}')
        t0+=h
        v0=vn
        error2.append(abs(analytic(t0)-v0))

rk2(2,25)


plt.plot([i for i in range(25)],error1,label='RK4')
plt.legend()
plt.title('Error RK4')
plt.show()
plt.plot([i for i in range(25)],error2,label='RK2')
plt.legend()
plt.title('Error RK2')
plt.show()
```
