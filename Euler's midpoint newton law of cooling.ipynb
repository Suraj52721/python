```python
import matplotlib.pyplot as plt
point1=[]
def euler(step):
    print('Eulers Method')
    ts=20
    t0=50
    k=0
    const=0.081
    for a in range(15):
        slope=-const*(t0-ts)
        t=t0+slope*step
        point1.append(t0)
        print(f't={k}m temp={t0}°C')
        k=k+step
        t0=t
        if t<0:
            break
euler(5)
print('\n')
plt.plot(point1,color='red',label='euler',marker='o')


point2=[]
def mid(step):
    print('MidPoint Method')
    ts=20
    t0=50
    k=0
    const=0.081
    slope=-const*(t0-ts)
    for a in range(15):
        t_mid=t0+slope*(step/2)
        slope=-const*(t_mid-ts)
        t=t0+slope*step
        point2.append(t0)
        print(f't={k}m  temp={t0}°C')
        k=k+step
        t0=t
mid(5)
print('\n')
plt.plot(point2,color='green',label='midpoint',marker='o')


point3=[]
def analytic():
    print('f(x)')
    ts=20
    t0=50
    const=0.081
    slope=-const*(t0-ts)
    for a in range(0,75,5):
        t=ts+(t0-ts)*(2.71)**(-const*a)
        print(f't={a}m temp={t}°C')
        point3.append(t)
analytic()
plt.plot(point3,label='f(x)',marker='o')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Temperature')
plt.show()
    
```
