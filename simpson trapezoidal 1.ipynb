```python
import matplotlib.pyplot as plt
import math
import numpy as np
def f(x):
  y=300*x/(1+(math.e)**x)
  return y

actual=246.5903
error1=[]
def trapezoidal(segment):
    for t in range(2,segment+2):
        a=0
        b=10
        sum=0
        list=np.linspace(a,b,t)
        for a0 in list:
            area=(a0-a)*(f(a)+f(a0))/2
            sum=sum+area
            a=a0
        error=abs(actual-sum)*100/actual
        error1.append(error)
        print(f'Area segment {t-1} ={sum}   error= {error}%')

error2=[]
def simpson(segment):
    for t in range(2,segment+2):
        a=0
        b=10
        sum=0
        list=np.linspace(a,b,t)
        for a0 in list:
            area=(a0-a)*(f(a)+4*f((a+a0)/2)+f(a0))/6
            sum=sum+area
            a=a0
        error=abs(actual-sum)*100/actual
        error2.append(error)
        print(f'Area segment {t-1} ={sum}   error= {error}%')

print('Trapezoidal Method')
trapezoidal(10)
print('\n')
print('Simpson Method')
simpson(10)
d=np.linspace(0,40,100)
plt.plot(f(d))
plt.xlim(0,30)
plt.title('f(x)')
plt.show()
plt.plot(error1,label='Trapezoidal',marker='o')
plt.plot(error2,label='Simpson',marker='o')
plt.title('Error')
plt.legend()
plt.show()
```
