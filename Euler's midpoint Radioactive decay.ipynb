```python
import matplotlib.pyplot as plt
import math
decay=0.0693
m0=100
point1=[]
def euler(step):
    print('Eulers Method')
    k=0
    m0=100
    for a in range(15):
        slope=-decay*m0
        m=m0+slope*step
        point1.append(m0)
        print(f't={k}s mass={m0}g')
        k=k+step
        m0=m
        if m<0:
            break
euler(10)
print('\n')
plt.plot(point1,color='red',label='euler',marker='o')


point2=[]
def mid(step):
    print('MidPoint Method')
    k=0
    m0=100
    slope=-decay*m0
    for a in range(15):
        m_mid=m0+slope*(step/2)
        slope=-decay*m_mid
        m=m0+slope*step
        point2.append(m0)
        print(f't={k}s  mass={m0}g')
        k=k+step
        m0=m
mid(10)
print('\n')
plt.plot(point2,color='green',label='midpoint',marker='o')


point3=[]
def analytic():
    print('f(x)')
    m0=100
    for a in range(0,150,10):
        m=m0*(2.71)**(-decay*a)
        print(f't={a}s mass={m}g')
        point3.append(m)
analytic()
plt.plot(point3,label='f(x)',marker='o')
plt.legend()
plt.xlabel('Time')
plt.ylabel('Mass')
plt.show()
    
```
