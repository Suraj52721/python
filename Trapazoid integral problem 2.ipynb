```python
import matplotlib.pyplot as plt
import math
import numpy as np
def f(x):
  y=300*x/(1+(math.e)**x)
  return y

actual=246.5903
def multiple(segment):
    for t in range(2,segment+2):
        a=0
        b=10
        sum=0
        list=np.linspace(a,b,t)
        for a0 in list:
            area=(a0-a)*(f(a)+f(a0))/2
            sum=sum+area
            a=a0
        error=abs(actual-sum)*100/actual
        print(f'Area segment {t-1} ={sum}   error= {error}%')
multiple(30)
d=np.linspace(0,40,100)
plt.plot(f(d))
plt.xlim(0,30)
plt.show()

def f(x):
    return np.sin(x)












```
