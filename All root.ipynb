```python
import matplotlib.pyplot as plt
import numpy as np
import math
def f(q):
  return q**3-50*q**2+336
tol=0.0001
def bisect(a,b,tol):
  if f(a)*f(b)<0:
    while (b-a)/2>tol:
      c=(a+b)/2
      if f(c)==0:
        print(c,'is the root.')
        break
      else:
        if f(a)*f(c)<0:
          b=c
        else:
          a=c
    print('the root is ',c)
    plt.plot(c, 0, marker="o", markersize=7 , 
             markeredgecolor="red",markerfacecolor="green")
    
  else:
    pass
    
x=np.linspace(-100,100,1000)
t=-100
for a in x:
  bisect(t,a,tol)
  t=a
m=np.linspace(-4,50,1000)
plt.plot(m,f(m))
plt.axhline()
plt.axvline()
plt.title('graph of x³-50x²+336')
plt.show()
```
