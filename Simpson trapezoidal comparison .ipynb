```python
import matplotlib.pyplot as plt
import math
import numpy as np
true_value = 246.5903
error_simpson = []
error_trapezoidal = []
def f(x):
    return 300*x/(1+(math.e)**x)
def simpson(a,b,segment):
    area = 0
    h = (b-a)/segment
    m=np.linspace(a,b,int(segment/2))
    for n in m:
        area+= h/3*(f(a)+4*f(a+h)+f(a+2*h))
        a+=2*h
       
    return area
def trapezoidal(a,b,segment):
    area = 0
    h = (b-a)/segment
    for n in range(0,segment):
        area += h/2*(f(a+n*h) + f(a+(n+1)*h))
    return area
def main():
    a = 0
    b = 10
    segment = 20
    for i in range(2,segment+2,2):
        print(f'{i} segments')
        print("Simpson's rule: ", simpson(a,b,i))
        print("Trapezoidal rule: ", trapezoidal(a,b,i))
        error_simpson.append(abs(true_value-simpson(a,b,i)))
        error_trapezoidal.append(abs(true_value-trapezoidal(a,b,i)))
main()
plt.plot(error_simpson,label='Simpson')
plt.plot(error_trapezoidal,label='trapedoidal')
plt.title('Error Graph')
plt.legend()
plt.xlabel('segment')
plt.ylabel('Error')
plt.show()
```
