```python
import math as m
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return np.tan(x)-x
def der(x):
    return 1/(np.cos(x)**2)-1

def newton(t,tol):
    for a in range(50):
        c=t-f(t)/der(t)
    
        if abs(t-c)>tol:
            t=c
            print('i=',a+1,'x0=',t,'(f(x0)',f(t))
        else:
            pass
    print('The Root is',c)
    plt.plot(c, 0, marker="o", markersize=7 , 
                 markeredgecolor="red",markerfacecolor="green")

def plot():
    x=np.linspace(-5,5,1000)
    plt.plot(x,f(x))
    plt.ylim(-5,5)
    plt.axhline()
    plt.axvline()
    plt.title('graph of tanx-x')
    plt.show()
newton(1,0.0001)
print('\n')
newton(4.7,0.0001)
print('\n')
newton(-4.6,0.0001)
plot()



```
