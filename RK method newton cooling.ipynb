```python
import math
import matplotlib.pyplot as plt
ts=20
t0=50
const=0.081
slope=-const*(t0-ts)

def f(t,t0):
  return -const*(t0-ts)

def analytic(t):
  return ts+(t0-ts)*(math.exp(-const*t))

error1=[]
def rk4(h,iteration):
  ts=20
  t0=50
  t=0
  print(f'RK4 Method')
  for a in range(iteration):
    k1=h*f(t,t0)
    k2=h*f(t+h/2,t0+k1/2)
    k3=h*f(t+h/2,t0+k2/2)
    k4=h*f(t+h,t0+k3)
    tn=t0+(k1+2*k2+2*k3+k4)/6
    print(f't={t} T={t0} analytic={analytic(t)}')
    t+=h
    t0=tn
    error1.append(abs(analytic(t)-tn))

rk4(5,15)

error2=[]
def rk2(h,iteration):
    ts=20
    t0=50
    t=0
    print(f'RK2 Method')
    for a in range(iteration):
        k1=h*f(t,t0)
        k2=h*f(t+h,t0+k1)
        tn=t0+(k1+k2)/2
        print(f't={t} T={t0} analytic={analytic(t)}')
        t+=h
        t0=tn
        error2.append(abs(analytic(t)-tn))

rk2(5,15)


plt.plot([i for i in range(15)],error1,label='RK4')
plt.legend()
plt.title('Error RK4')
plt.show()
plt.plot([i for i in range(15)],error2,label='RK2')
plt.legend()
plt.title('Error RK2')
plt.show()
```
