```python
import matplotlib.pyplot as plt
import numpy as np
p=[0.94,1.05,1.17,1.28]
T=[0,30,70,100]
pT_xy=[]
for a in range(len(p)):
  sum=T[a]*p[a]
  pT_xy.append(sum)
print(f'pT_xy = {pT_xy}')
Sx=0
Sy=0
Sxy=0
Sxx=0
n=len(p)
for i in range(n):
  Sx+=T[i]
  Sy+=p[i]
  Sxy+=pT_xy[i]
  Sxx+=(T[i])**2
print(f'Sx = {Sx}')
print(f'Sy = {Sy}')
print(f'Sxy = {Sxy}')
print(f'Sxx = {Sxx}')
a1= (n*Sxy-Sx*Sy)/(n*Sxx-(Sx)**2)
print(f'a1 = {a1}')
a0=(Sxx*Sy-Sxy*Sx)/(n*Sxx-(Sx)**2)
print(f'a0 = {a0}')
for i in range(4):
  plt.plot(p[i],T[i],marker='o')
x=np.linspace(0,100,100)
y=a0+a1*x
plt.plot(y,x)
plt.show()
```
