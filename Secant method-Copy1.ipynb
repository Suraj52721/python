```python

import math as m
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return np.tan(x)-x

def secant(a,b,tol):
    for m in range(50):
        if abs(a-b)/2>tol:
            c=a-(f(a)*(a-b))/(f(a)-f(b))
            a=b
            b=c
            print('i=',m+1,'x0=',c,'f(a)=',f(a),'f(b)=',f(b))
        else:
            pass
    print('The root is ',c)
    plt.plot(c, 0, marker="o", markersize=7 , 
             markeredgecolor="red",markerfacecolor="green")
secant(-4,-4.5,0.0001)
print('\n')
secant(0,1,0.0001)
print('\n')
secant(7,7.5,0.0001)
x=np.linspace(-10,10,1000)
plt.plot(x,f(x))
plt.axhline()
plt.axvline()
plt.ylim(-5,5)
plt.title('graph of xÂ³-5x-1')
plt.show()

```
